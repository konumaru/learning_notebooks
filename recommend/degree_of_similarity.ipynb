```python
import itertools
import pandas as pd
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

## データセットの作成

```python
dataset = {
    'A': [2.5, 3.5, 3.0, 3.5, 2.5, 3.0],
    'B': [3.0, 3.5, 1.5, 5.0, 3.0, 3.5],
    'C': [2.5, 3.0, None, 3.5, None, 4.0],
    'D': [None, 3.5, 3.0, 4.0, 2.5, 4.5],
    'E': [3.0, 4.0, 2.0, 3.0, 2.0, 3.0],
    'F': [3.0, 4.0, None, 5.0, 3.5, 3.0],
    'G': [None, 4.5, None, 4.0, 1.0, None]
}
col = ['カレー', 'ラーメン', 'チャーハン', '寿司', '牛丼', 'うどん']
```

```python
origin_df = pd.DataFrame(dataset)
origin_df.index = col
```

## ユークリッド距離

それぞれa, bの要素 $$(a_1, a_2, a_3,....a_i),(b_1, b_2, b_3,....b_i)$$
ab間のユークリッド距離dは $$ d = \sqrt{(a_1-b_1)^2+(a_2-b_2)^2+(a_3-b_3)^2+.....+(a_i-b_i)^2} $$

```python
lebels = origin_df.copy().columns
df = pd.DataFrame(index=lebels, columns=lebels)
```

```python
def cal_score(row):
    p1, p2 = row
    return pow(p1-p2, 2)

for persons in list(itertools.permutations(df.columns, 2)):
    cal_df = origin_df[[persons[0], persons[1]]] # 元データから必要な値を取り出す
    cal_df = cal_df.dropna(how='any', axis=0) # 欠損のある値を削除
    df.loc[persons[0], persons[1]] = np.sqrt(sum(cal_df.apply(cal_score, axis=1))) # ユークリッド距離の計算

# 値を0〜1に揃える
df = df / (1+df)
df.fillna(1, inplace=True)
```

```python
plt.figure(figsize=(9, 6))
sns.heatmap(df, annot=True, fmt='g', cmap='Blues')
```

## ピアソン相関

$$
\frac{\sum_{i=1}^{n} (x_i - \bar{x})(y_i - \bar{y})}
{\sqrt{\sum_{i=1}^{n} (x_i - \bar{x})^2}\sqrt{\sum_{i=1}^{n} (y_i - \bar{y})^2}}
$$
